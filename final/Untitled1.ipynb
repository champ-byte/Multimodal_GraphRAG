```python
from transformers import pipeline

# Load the model as a text2text generator
extractor = pipeline("text2text-generation", model="Babelscape/rebel-large")

text = "A kettle sat on a wooden table next to a rusty lamp."

# Generate triples
output = extractor(text, max_length=512, truncation=True)

print(output)
```
